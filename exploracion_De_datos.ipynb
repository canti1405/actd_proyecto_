```python
from openpyxl import load_workbook
import pandas as pd

# Cargar el libro de Excel sin leer los datos todavía
excel_path = "datos_utiles.xlsx"
workbook = load_workbook(excel_path, read_only=True)

# Mostrar los nombres de las hojas disponibles
sheet_names = workbook.sheetnames
sheet_names
df = pd.read_excel(excel_path)
```

```python
# Generar estadísticas descriptivas
desc_stats = df.describe().T
desc_stats
```

### Histogramas

```python
import matplotlib.pyplot as plt

# Histogramas de variables relevantes
variables = ['edad', 'fami_estratovivienda', 'punt_global']

for col in variables:
    plt.figure(figsize=(6,4))
    df[col].hist(bins=30, edgecolor='black')
    plt.title(f'Histograma de {col}')
    plt.xlabel(col)
    plt.ylabel('Frecuencia')
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
import seaborn as sns

# Boxplot: puntaje global por estrato
plt.figure(figsize=(8,5))
sns.boxplot(x='fami_estratovivienda', y='punt_global', data=df)
plt.title('Boxplot de Puntaje Global por Estrato')
plt.xlabel('Estrato')
plt.ylabel('Puntaje Global')
plt.tight_layout()
plt.show()
```

```python
# Dispersión entre edad y puntaje global
plt.figure(figsize=(7,5))
sns.scatterplot(x='edad', y='punt_global', data=df, alpha=0.5)
plt.title('Dispersión entre Edad y Puntaje Global')
plt.xlabel('Edad')
plt.ylabel('Puntaje Global')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Diagrama de violín para comparar puntaje global según si tiene computador
plt.figure(figsize=(8,5))
sns.violinplot(x='fami_tienecomputador', y='punt_global', data=df)
plt.title('Diagrama de Violín: Puntaje Global según Tenencia de Computador')
plt.xlabel('Tiene Computador (0 = No, 1 = Sí)')
plt.ylabel('Puntaje Global')
plt.tight_layout()
plt.show()
```
